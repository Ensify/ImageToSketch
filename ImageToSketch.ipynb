```python
def imgToSketch(img,disp=False,show_orig = False,edge=False,thres=100):
    image = cv2.imread(img)
    if show_orig: cv2_imshow(image)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray,100,200)
    if edge: cv2_imshow(255-edges)
    _, shade = cv2.threshold(gray, thres, 255, cv2.THRESH_BINARY)
    shade[shade<100] = 50
    sketch = cv2.subtract(shade,edges)
    if disp: cv2_imshow(sketch)
    return sketch
```

```python
sktch = imgToSketch("sample2.png",True,True)
```
